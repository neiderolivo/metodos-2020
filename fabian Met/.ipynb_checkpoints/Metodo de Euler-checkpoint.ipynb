Problema del Paracadidista:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from pylab import plot, show, grid
from math import exp
```

```python
g=9.81
m=68.1
c=12.5

def dvdt(t,v):
    return (c/m)*v-g
```

```python
n=20
to=0
tf=20
vo=0
h=(tf-to)/n
t=[to]
v=[vo]

for i in range(n):
    vo=vo+h*dvdt(to,vo)
    v.append(vo)
    to=to+h
    t.append(to)

```

```python
tm=np.arange(0,20,0.1)
Vana=53.3904*(1-2.7183**(0.184*tm))
plot(tm,Vana)
plot(t,v)
plt.title('Grafica analítica y Numérica')
plt.xlabel('Tiempo')
plt.ylabel('Velocidad')
grid("on")
show()
```
